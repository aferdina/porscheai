```python
import numpy as np

ref_secs = [0, 2, 4, 5, 7, 10]
ref_speeds = [5, 2, 15, 15, 10, 8]
sim_stepsize = 0.01 #Seconds

sim_duration = ref_secs[-1]
total_steps = sim_duration/sim_stepsize

ref_trajectory = np.array([])

for idx in range(len(ref_secs)-1):
    sect_dur = ref_secs[idx+1] - ref_secs[idx]
    curr_speed = ref_speeds[idx]
    next_speed = ref_speeds[idx+1]
    num_steps = int(total_steps * sect_dur/sim_duration)

    section = np.linspace(curr_speed, next_speed, num_steps)
    ref_trajectory = np.append(ref_trajectory, section)
```

```python
import matplotlib.pyplot as plt
x_ticks = np.linspace(0, sim_duration, int(total_steps))
plt.plot(x_ticks, ref_trajectory)
plt.xlabel("Time (Sec.)")
plt.ylabel("Speed (m/s)")
plt.title("Reference Trajectory")
plt.grid()
plt.savefig("Reference Trajectory")
```
